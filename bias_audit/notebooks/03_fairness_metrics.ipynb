# 03 - Fairness metrics

Demographic parity, equalized odds, TPR/FPR, predictive parity et métriques par groupe.

```python
import json
import pandas as pd
from pathlib import Path
metrics_path = Path('../results/metrics/audit_metrics.json')
if not metrics_path.exists():
    raise FileNotFoundError('Lancez scripts/run_all.sh ou scripts/evaluate_fairness.py avant ce notebook.')
data = json.loads(metrics_path.read_text(encoding='utf-8'))
rows = []
for attr, fairness in data['baseline_evaluation']['fairness'].items():
    for group, vals in fairness['group_metrics'].items():
        rows.append({'attribute': attr, 'group': group, **vals})
pd.DataFrame(rows)
```
