Objectif : charger COMPAS, appliquer le nettoyage projet et inspecter les distributions des attributs sensibles, du label et des base rates.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path('..').resolve()))
import pandas as pd
from src.data_processing import prepare_compas_data, DataProcessor

raw = pd.read_csv('../data/raw/compas-scores-two-years.csv')
df = prepare_compas_data(raw)
processor = DataProcessor(['race', 'sex'], 'two_year_recid')
display(df.head())
display(pd.DataFrame(processor.explore_demographics(df)['race']['counts'], index=['count']).T)
display(df.groupby('race')['two_year_recid'].mean().rename('base_rate'))
display(df.groupby('sex')['two_year_recid'].mean().rename('base_rate'))
```
